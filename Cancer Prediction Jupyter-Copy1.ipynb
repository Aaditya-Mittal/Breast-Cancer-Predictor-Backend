```python
import pickle
def FitModel (X,Y, algo_name , algorithm, gridSearchParams, cv):
    np.random.seed(10)
    x_train, x_test, y_train, y_test = train_test_split (X,Y,test_size = 0.2)

    grid = GridSearchCV(estimator = algorithm, param_grid = gridSearchParams,
                        cv = cv, scoring = 'accuracy', verbose = 1 , n_jobs = -1 )
    
    grid_result = grid.fit(x_train, y_train)
    best_params = grid_result.best_params_
    
    x_test_clean = np.ascontiguousarray(x_test.values if hasattr(x_test, 'values') else x_test, dtype=np.float64)
    
    pred = grid_result.predict (x_test_clean)
    cm = confusion_matrix (y_test,pred)
    print (pred)
    pickle.dump(grid_result,open(algo_name,'wb'))
    
    print ('Best Params :', best_params)
    print ('Classification Report:',classification_report(y_test,pred))
    print ('Accuracy Score', (accuracy_score(y_test,pred)))
    print ('Confusion Matrix :\n',cm)

```

```python
#Starting with a simple model of logistic regression
from sklearn.linear_model import LogisticRegression
params ={'C':[0.5,0.8,1.0,1.2,1.5]}
FitModel (x,y, 'Logistic Regression' , LogisticRegression(), params, 10)
```

```python
#Trying to find a better C value for the model. Increased accuracy by 1%
params ={'C':[1.1,1.2,1.3]}
FitModel (x,y, 'Logistic Regression' , LogisticRegression(), params, 10)
```

```python
#using svm now
from sklearn.svm import SVC
param = {'C': [0.1,1,100,1000], 'gamma':[0.0001,0.001, 0.005, 0.1,1, 3,5,10, 100]  }
FitModel (x,y,'SVC',SVC(), param, cv =10)
```

```python
#trying to optimise params for svm, but no incease in accuracy here
param = {'C': [0.8,0.9,1.0,1.1,1.2], 'gamma':[0.0009,0.001, 0.002, 0.003]  }
FitModel (x,y,'SVC',SVC(), param, cv =10)
```

```python
#using random forest
from sklearn.ensemble import RandomForestClassifier
param = { 'n_estimators': [100,500,1000,2000]  }
FitModel (x,y,'Random Forest',RandomForestClassifier(), param, cv =10)
```

```python
#optimising random forest params no improvement in accuracy
param = { 'n_estimators': [400,450,500,550,600]  }
FitModel (x,y,'Random Forest High Accuracy',RandomForestClassifier(), param, cv =10)
```

```python
#using decision tree now
from sklearn.tree import DecisionTreeClassifier
param = { 'min_samples_split': [2,5,8,10,12,15]  }
FitModel (x,y,'Decision Tree',DecisionTreeClassifier(), param, cv =10)
```

```python
#optimising DECISION TREE params but no effect on accuracy
param = { 'min_samples_split': [3,4,5,6,7]  }
FitModel (x,y,'Decision Tree',DecisionTreeClassifier(), param, cv =10)
```

```python
#using knn now
from sklearn.neighbors import KNeighborsClassifier
param = { 'n_neighbors': [5,10,15,20]  }
FitModel (x,y,'KNN',KNeighborsClassifier(), param, cv =10)
```

```python
#optimising KNN params but no effect on accuracy
param = { 'n_neighbors': [16,19,20,21]  }
FitModel (x,y,'KNN',KNeighborsClassifier(), param, cv =10)
```

```python
#using XGBOOST now
from xgboost import XGBClassifier
param = { 'n_estimators': [100,500,1000,2000]  }
FitModel (x,y,'XGBoost', XGBClassifier(),param, cv = 10)
```

```python
#optimising xgboost params but no effect on accuracy
param = { 'n_estimators': [100,200,300,400]  }
FitModel (x,y,'XGBoost', XGBClassifier(),param, cv = 10)
```
